```python
import json
import os
from pathlib import Path

import requests
from pandas import DataFrame
```

```python
ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
ROOT_DIR
```

```python
url = 'https://www.overview.hk/street/ssp.php'

response = requests.get(url)
raw_data = json.loads(response.text)
raw_data
```

```python
info = {
    'chinese_road_name': [],
    'chinese_district_name': [],
    'english_road_name': [],
    'english_district_name': []
}

for detail in raw_data['data']:
    info['chinese_road_name'].append(detail[0])
    info['chinese_district_name'].append(detail[2])
    info['english_road_name'].append(detail[1])
    info['english_district_name'].append(detail[3])
```

```python
dataframe = DataFrame(info)
dataframe   
```

```python
dataframe.to_csv(f'{ROOT_DIR}/data/raw/road_info.csv', index=False)
```
