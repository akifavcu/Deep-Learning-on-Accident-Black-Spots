```python
import os
from pathlib import Path

import pandas
```

```python
ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
ROOT_DIR
```

```python
dataframe = pandas.read_csv(f'{ROOT_DIR}/data/raw/osm_edge_info.csv')
dataframe
```

```python
base_edge_list = []

for _, info in dataframe.iterrows():
    start_node_id, end_node_id = info['node2node_id'].lstrip('(').strip(')').split(', ')[:2]
    base_edge_list.append(f'{start_node_id} {end_node_id}\n')

base_edge_list
```

```python
with open(f'{ROOT_DIR}/data/processed/node2vec/road_structure_base.el', 'w') as file:
    file.writelines(base_edge_list)
```

```python
lanes_edge_list = []

for _, info in dataframe.iterrows():
    start_node_id, end_node_id = info['node2node_id'].lstrip('(').strip(')').split(', ')[:2]
    lanes = info['lanes'] if str(info['lanes']) else 1.0
    lanes_edge_list.append(f'{start_node_id} {end_node_id} {lanes}\n')

lanes_edge_list
```

```python
with open(f'{ROOT_DIR}/data/processed/node2vec/road_structure_lanes.el', 'w') as file:
    file.writelines(lanes_edge_list)
```

```python
length_edge_list = []

for _, info in dataframe.iterrows():
    start_node_id, end_node_id = info['node2node_id'].lstrip('(').strip(')').split(', ')[:2]
    length = info['length']
    length_edge_list.append(f'{start_node_id} {end_node_id} {length}\n')

length_edge_list
```

```python
with open(f'{ROOT_DIR}/data/processed/node2vec/road_structure_length.el', 'w') as file:
    file.writelines(length_edge_list)
```

```python
ref_edge_list = []

for _, info in dataframe.iterrows():
    start_node_id, end_node_id = info['node2node_id'].lstrip('(').strip(')').split(', ')[:2]
    ref = info['ref'] if str(info['ref']) != 'nan' else 0.0
    ref_edge_list.append(f'{start_node_id} {end_node_id} {ref}\n')

ref_edge_list
```

```python
with open(f'{ROOT_DIR}/data/processed/node2vec/road_structure_ref.el', 'w') as file:
    file.writelines(length_edge_list)
```

```python
maxspeed_edge_list = []

for _, info in dataframe.iterrows():
    start_node_id, end_node_id = info['node2node_id'].lstrip('(').strip(')').split(', ')[:2]
    maxspeed = info['maxspeed'] if str(info['maxspeed']) != 'nan' else 50.0
    maxspeed_edge_list.append(f'{start_node_id} {end_node_id} {maxspeed}\n')

maxspeed_edge_list
```

```python
with open(f'{ROOT_DIR}/data/processed/node2vec/road_structure_maxspeed.el', 'w') as file:
    file.writelines(maxspeed_edge_list)
```
