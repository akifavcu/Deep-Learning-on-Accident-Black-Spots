```python
import os
from pathlib import Path

import pandas
```

```python
ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
ROOT_DIR
```

```python
dataframe = pandas.read_csv(f'{ROOT_DIR}/data/interim/road_structure_angle_info.csv')
dataframe
```

```python
angle_list = []

for _, info in dataframe.iterrows():
    start_node_id = info['start_node_id']
    end_node_id = info['end_node_id']
    road_angle = info['road_angle']
    angle_list.append(f'{start_node_id} {end_node_id} {road_angle}\n')

angle_list
```

```python
with open(f'{ROOT_DIR}/data/processed/node2vec/road_structure_angle.el', 'w') as file:
    file.writelines(angle_list)
```
