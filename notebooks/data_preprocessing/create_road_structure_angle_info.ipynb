```python
import math
import os
from pathlib import Path

import pandas
from pandas import DataFrame
```

```python
ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
ROOT_DIR
```

```python
node_dataframe = pandas.read_csv(f'{ROOT_DIR}/data/raw/osm_node_info.csv')
node_dataframe
```

```python
node_info_dict = {}

for _, node_info in node_dataframe.iterrows():
    node_info_dict[node_info['node_id']] = {'x': node_info['x'], 'y': node_info['y']}

print(len(node_info_dict))
node_info_dict
```

```python
edge_dataframe = pandas.read_csv(f'{ROOT_DIR}/data/raw/osm_edge_info.csv')
edge_dataframe
```

```python
edge_info_dict = {}

for _, edge_info in edge_dataframe.iterrows():
    start_node_id, end_node_id = edge_info['node2node_id'].lstrip('(').strip(')').split(', ')[:2]
    if start_node_id not in edge_info_dict:
        edge_info_dict[start_node_id] = [end_node_id]
    else:
        edge_info_dict[start_node_id].append(end_node_id)
    if end_node_id not in edge_info_dict:
        edge_info_dict[end_node_id] = [start_node_id]
    else:
        edge_info_dict[end_node_id].append(start_node_id)

edge_info_dict
```

```python
def calculate_initial_compass_bearing(start_lat_long, end_lat_long):
    start_lat = math.radians(start_lat_long[0])
    end_lat = math.radians(end_lat_long[0])
    diff_long = math.radians(end_lat_long[1] - start_lat_long[1])
    x = math.sin(diff_long) * math.cos(end_lat)
    y = math.cos(start_lat) * math.sin(end_lat) - (math.sin(start_lat) * math.cos(end_lat) * math.cos(diff_long))
    initial_bearing = math.degrees(math.atan2(x, y))
    compass_bearing = (initial_bearing + 360) % 360
    return compass_bearing
```

```python
def calculate_angle(start_lat_long, mid_lat_long, end_lat_long):
    start_mid_bearing = calculate_initial_compass_bearing(start_lat_long, mid_lat_long)
    mid_end_bearing = calculate_initial_compass_bearing(mid_lat_long, end_lat_long)
    angle = abs(mid_end_bearing - start_mid_bearing)
    if angle > 180:
        angle = 360 - angle
    return angle
```

```python
road_structure_angle_info = {
    'start_node_id': [],
    'mid_node_id': [],
    'end_node_id': [],
    'road_angle': []
}

for start_node_id in edge_info_dict:
    for mid_node_id in edge_info_dict[start_node_id]:
        for end_node_id in edge_info_dict[mid_node_id]:
            start_node_id = int(start_node_id)
            mid_node_id = int(mid_node_id)
            end_node_id = int(end_node_id)
            if start_node_id in node_info_dict and mid_node_id in node_info_dict and end_node_id in node_info_dict:
                start_node_lat_long = (node_info_dict[start_node_id]['y'], node_info_dict[start_node_id]['x'])
                mid_node_lat_long = (node_info_dict[mid_node_id]['y'], node_info_dict[mid_node_id]['x'])
                end_node_lat_long = (node_info_dict[end_node_id]['y'], node_info_dict[end_node_id]['x'])
                road_structure_angle_info['start_node_id'].append(start_node_id)
                road_structure_angle_info['mid_node_id'].append(mid_node_id)
                road_structure_angle_info['end_node_id'].append(end_node_id)
                road_angle = calculate_angle(start_node_lat_long, mid_node_lat_long, end_node_lat_long)
                road_structure_angle_info['road_angle'].append(road_angle)
```

```python
road_structure_angle_dataframe = DataFrame(road_structure_angle_info)
road_structure_angle_dataframe
```

```python
road_structure_angle_dataframe.to_csv(f'{ROOT_DIR}/data/interim/road_structure_angle_info.csv', index=False)
```
